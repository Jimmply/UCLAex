# Uni-, Bi-, and Multi-Variate Data Visualization with Python

Execute the following cell to import libraries and load the datasets.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mpg_data = sns.load_dataset('mpg')
diamond_data = sns.load_dataset('diamonds')
```

## Univariate and Bivariate

We now have a variable called `mpg_data` that is a Pandas dataframe storing data about cars.  Execute the following cell to look at the first five records.

```python
mpg_data.head()
```

Make a bar plot of data in the `cylinders` column showing the count of each cylinder value.  When doing this:
* order the bars from highest count to lowest count
* label the horizontal and vertical axes
* add a title
* make the bars blue
* change the font sizes of the labels and title

```python
cylinder_counts = mpg_data['cylinders'].value_counts().sort_values(ascending=False)


sns.barplot(
    x=cylinder_counts.index, 
    y=cylinder_counts.values, 
    color='blue',
    order=cylinder_counts.index
)

plt.xlabel('Number of Cylinders', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.title('Count of Cars by Cylinder', fontsize=16)
plt.show()
```

Make another bar plot of data in the `cylinders` column showing the count of each cylinder value.  Make it the same as the plot above, only now order the bars based on the cylinder values.

```python
cylinder_counts = mpg_data['cylinders'].value_counts().sort_index()

sns.barplot(
    x=cylinder_counts.index, 
    y=cylinder_counts.values, 
    color='blue',
    order=cylinder_counts.index
)

plt.xlabel('Number of Cylinders', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.title('Count of Cars by Cylinder', fontsize=16)
plt.show()
```

Make a histogram plot of data in the `mpg` column.  When doing this:
* make 10 total bins for the `mpg` values
* label the horizontal and vertical axes
* add a title
* make the bars blue with a black edge color 
  * *hint: you may find it useful to use the "edgecolor" parameter*
* change the font sizes of the labels and title

```python
plt.hist(
    mpg_data['mpg'], 
    bins=10,               
    color='blue',         
    edgecolor='black'     
)

plt.xlabel('Miles per Gallon (mpg)', fontsize=12)
plt.ylabel('Count of Cars', fontsize=12)
plt.title('Distribution of Miles per Gallon', fontsize=16)
plt.show()
```

Make another histogram plot of data in the `mpg` column.  Make it the same as the plot above, only now include 30 bins for the `mpg` values.

```python
plt.hist(
    mpg_data['mpg'], 
    bins=30,               
    color='blue',         
    edgecolor='black'     
)

plt.xlabel('Miles per Gallon (mpg)', fontsize=12)
plt.ylabel('Count of Cars', fontsize=12)
plt.title('Distribution of Miles per Gallon', fontsize=16)
plt.show()
```

Make a scatter plot of `mpg` versus `horsepower`.  When doing this:
* make sure the horizontal and vertical axes are labelled
* add a title
* change the tick marks so that the mpg axis only has ticks for 10 and 45 and the horsepower axis only has ticks for 50 and 225
* make the markers green
* change the font sizes of the labels and title

```python
plt.scatter(
    x=mpg_data['horsepower'], 
    y=mpg_data['mpg'], 
    color='green'
)

plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Miles per Gallon (mpg)', fontsize=12)
plt.title('MPG vs. Horsepower', fontsize=16)
plt.xticks([50, 225])
plt.yticks([10, 45])
plt.show()
```

Make a box plot of `mpg` vs `cylinders`.  When doing this:
* the figure should show one box for each cylinder value, where the box shows the distribution of mpg values
* make the boxes orange
* make sure the horizontal and vertical axes are labelled
* add a title
* change the font sizes of the labels and title

```python
sns.boxplot(
    x='cylinders', 
    y='mpg', 
    data=mpg_data, 
    color='orange'
)

plt.xlabel('Number of Cylinders', fontsize=12)
plt.ylabel('Miles per Gallon (mpg)', fontsize=12)
plt.title('MPG vs Cylinders', fontsize=16)
plt.show()
```

## Multivariate

`diamond_data` is a Pandas dataframe containing data related to diamonds.  We are going to make several multivariate visualizations with it.

Execute the following cell to look at the first five records.

```python
diamond_data.head()
```

Make a pair plot of data in the `price`, `carat`, `x`, `y`, and `z` columns, with colors corresponding to the `cut` column values.  This plot should look like the following:

<img src='https://raw.githubusercontent.com/benjum/UCLAX-25F-EDA/main/Images/assignment-images/PairplotDiamonds.png' width=500px>

```python
cols_to_plot = ['price', 'carat', 'x', 'y', 'z']

sns.pairplot(
    diamond_data[cols_to_plot + ['cut']],
    hue='cut',
)
plt.show()
```

Make a box plot for `carat` vertically and `cut` horizontally, with colors corresponding to `color` column values.  This plot should look like the following:

<img src='https://raw.githubusercontent.com/benjum/UCLAX-25F-EDA/main/Images/assignment-images/Box1diamonds.png' width=250px>

```python
sns.boxplot(
    x='cut', 
    y='carat', 
    hue='color',
    data=diamond_data
)
plt.legend(title='Color', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show
```

Make a box plot for `price` vertically and `cut` horizontally, with colors corresponding to `color` column values.  This plot should look like the following:

<img src='https://raw.githubusercontent.com/benjum/UCLAX-25F-EDA/main/Images/assignment-images/Box2diamonds.png' width=250px>

```python
sns.boxplot(
    x='cut', 
    y='price', 
    hue='color',
    data=diamond_data
)
plt.legend(title='Color', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show
```

Make a scatter plot with `price` vertically and `carat` horizontally, with colors corresponding to `cut` column values.  This plot should look like the following:

<img src='https://raw.githubusercontent.com/benjum/UCLAX-25F-EDA/main/Images/assignment-images/ScatterDiamonds.png' width=500px>

```python
sns.scatterplot(
    x='carat', 
    y='price', 
    hue='cut',   
    data=diamond_data
)


plt.legend(title='Cut', loc='lower right')
plt.show()
```

Make a grid of scatter plots with `price` vertically and `carat` horizontally, with colors corresponding to `cut` column values, and with each plot in the grid only plotting data from a unique value of `cut`.  This plot should look like the following:

<img src='https://raw.githubusercontent.com/benjum/UCLAX-25F-EDA/main/Images/assignment-images/ScatterGridDiamonds.png' width=500px>

```python
g = sns.FacetGrid(
    diamond_data, 
    col='cut',
    col_wrap=3
)

g.map_dataframe(
    sns.scatterplot, 
    x='carat', 
    y='price', 
    hue='cut'
)

plt.show()
```
